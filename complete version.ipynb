```python
stage=0
result=0
bcount=1
dataname = 'Subject ' + str(subject) + '.csv'
seen = []
mouse_pos = []
mouse_click = []
root = tk.Tk()
root.title("My Experiment")
root.geometry("1280x1024")
root.minsize(1280,1024)
root.configure(background="white")
consent = "Dear participant, thank you very much for taking part in our study. \n \n This experiment consists of three blocks presentation of ninety contexts and images. \n You need to reply questionaires after the presentation of each context. \n\n Please pay attention during the presentation of contexts and images.  \nYou need to imagine the written situation in order to choose between images. \n After the presentation of go signal that is a pop up  black dot, \nclick left button of the mouse if you want to select the left image, \nor click right button if you want to the right image. \n\n You should reply the questions carefully and honestly. \n\n If you have any questions regarding the study, please contact the experimenter \nbefore the beginning of the study or during the break between blocks before \nstarting to a new block. "

title = tk.Label(text=consent, wraplength = 1000, background ="white", font=("Times",16))
title.place(relx=0.5, rely=0.3, anchor="center")


def show_context():
    # get random context
    global display_context
    context = df["Contexts"][get_rand()]
    display_context = tk.Label(master=root, text=context, height=30, bg="white", font=("None", 16))
    display_context.place(relx=0.5, rely=0.5, anchor="center")
    print("Stage " + str(get_stage()) + " Context " + str(result))

def hide_context():
    global display_context
    display_context.place_forget()

def get_rand():
    global result
    result = random.randint(0,89)
    if result not in seen:
        seen.append(result)
        return result
    else:
        while result in seen:
            result = random.randint(0,89)
            if result not in seen:
                seen.append(result)
                return result
                    
def show_images():
    # get relevant images
    global display_image1, display_image2
    image1 = df["Ph A"][result]
    image2 = df["Ph B"][result] 
    img = ImageTk.PhotoImage(Image.open(image1))
    img2 = ImageTk.PhotoImage(Image.open(image2))
    display_image1 = tk.Label(root, image=img)
    display_image2 = tk.Label(root, image=img2)
    display_image1.image=img
    display_image2.image=img2
    display_image1.place(relx=0.2, rely=0.5, anchor="center")
    display_image2.place(relx=0.8, rely=0.5, anchor="center")

def hide_images():
    display_image1.place_forget()
    display_image2.place_forget()


def show_cross():
    global display_cross
    cross = "x2cross.png"
    cross_img=ImageTk.PhotoImage(Image.open(cross))
    display_cross= tk.Label(root, image=cross_img, borderwidth=0,bg="white")
    display_cross.image=cross_img
    display_cross.place(relx=0.5,rely=0.5, anchor="center")

def hide_cross():
    global display_cross
    display_cross.place_forget()

def show_go():
    global display_go
    go = "go2.jpg"
    go_img = ImageTk.PhotoImage(Image.open(go))
    display_go = tk.Label(root, image=go_img, borderwidth=0)
    display_go.image=go_img
    display_go.place(relx=0.5, rely=0.5, anchor="center")
    
def hide_go():
    global display_go
    display_go.place_forget()    
    
    
def show_empty():
    global display_empty
    empty = "go.jpg"
    empty_img = ImageTk.PhotoImage(Image.open(empty))
    display_empty = tk.Label(root, image=empty_img, borderwidth=0)
    display_empty.image=empty_img
    display_empty.place(relx=0.5, rely=0.5, anchor="center")
    
def hide_empty():
    global display_empty, mouse_click
    display_empty.place_forget()
    mouse_click=[]
    
def show_decision():
    global display_dec
    dec = "go.jpg"
    dec_img = ImageTk.PhotoImage(Image.open(dec))
    display_dec = tk.Label(root, image=dec_img, borderwidth=0)
    display_dec.image=dec_img
    display_dec.place(relx=0.5, rely=0.5, anchor="center")
    
def hide_decision():
    global display_dec, mouse_click
    display_dec.place_forget()           
    saveitem(mouse_click)


def show_complex():
    
    global question, scale, info1
    
    question = tk.Label(text="How complex was the image you chose?", wraplength = 500, background ="white", font=("Times",20))
    question.place(relx=0.5, rely=0.5, anchor="center")
    info1 = tk.Label(text="(-10: not complex at all, 10: very complex)", background ="white", font=("Times",15))
    info1.place(relx = 0.5, rely = 0.56, anchor ="center")
    scale = Scale(root, from_=-10, to=10, orient=HORIZONTAL,  resolution=1, length= 400, width = 30, repeatdelay = 0, command = get_complex)
    scale.place(relx = 0.5, rely = 0.7, anchor ="center")
    
def hide_complex():
    global question, scale, info1
    question.place_forget()   
    scale.place_forget()
    info1.place_forget()
    
def show_interesting():
    
    global question2, scale2, info2
    
    question2 = tk.Label(text="How interesting was the image you chose?", wraplength = 500, background ="white", font=("Times",20))
    question2.place(relx=0.5, rely=0.5, anchor="center")
    info2 = tk.Label(text="(-10: not interesting at all, 10: very interesting)", background ="white", font=("Times",15))
    info2.place(relx = 0.5, rely = 0.56, anchor ="center")
    scale2 = Scale(root, from_=-10, to=10, orient=HORIZONTAL,  resolution=1, length= 400, width = 30, repeatdelay = 0, command = get_interesting)
    scale2.place(relx = 0.5, rely = 0.7, anchor ="center")
    
def hide_interesting():
    global question2, scale2, info2
    question2.place_forget()   
    scale2.place_forget()
    info2.place_forget()    
    
def show_rewarding():
    
    global question3, scale3, info3
    
    question3 = tk.Label(text="How rewarding was the image you chose?", wraplength = 500, background ="white", font=("Times",20))
    question3.place(relx=0.5, rely=0.5, anchor="center")
    info3 = tk.Label(text="(-10: not rewarding at all, 10: very rewarding)", background ="white", font=("Times",15))
    info3.place(relx = 0.5, rely = 0.56, anchor ="center")
    scale3 = Scale(root, from_=-10, to=10, orient=HORIZONTAL,  resolution=1, length= 400, width = 30, repeatdelay = 0, command = get_rewarding)
    scale3.place(relx = 0.5, rely = 0.7, anchor ="center")
    

def breaktime():
    global wtext, contx, bcount

    if bcount ==3:
        thanks = tk.Label (text="Thank you for your participation. You will soon be debriefed by the experimenter.", background ="white", font=("Times",16))
        thanks.place(relx=0.5, rely=0.5, anchor='center')
        df.to_csv(dataname)

    else:
        contx = tk.Button(text="Continue", command=breakend, bg="white")
        contx.place(relx=0.5, rely=0.5,anchor="center")
        wtext = tk.Label(text="Break Time! Press continue when you are ready.", background="white", font =("Times",16))
        wtext.place(relx=0.5, rely=0.4,anchor="center")

def breakend():
    global wtext, contx, bcount
    bcount +=1
    wtext.place_forget()
    contx.place_forget()
    exp()
    
def hide_rewarding():
    global question3, scale3, info3
    question3.place_forget()   
    scale3.place_forget()
    info3.place_forget()    
    
def get_complex():
    
    df.loc[result ,'Complex'] = scale.get()    

def get_interesting():

    df.loc[result ,'Interesting'] = scale2.get()
    
def get_rewarding():      

    df.loc[result ,'Rewarding'] = scale3.get()

    
def save_cursor():
    win32gui.GetCursorPos()
    flags, hcursor, (x,y) = win32gui.GetCursorInfo()
    cursor=[]
    for i in range (10000):
        flags, hcursor, (x,y) = win32gui.GetCursorInfo()
        z = ([x,y])
        cursor.append(z)

def hidecursor():
    root.config(cursor='none')
    
def showcursor():
    root.config(cursor='')
    
def update_clock():
    global now
    now = datetime.now().strftime("%H:%M:%S.%f")
    #now = time.strftime("%H:%M:%S")
    root.after(1000, update_clock)
    return now

def get_time():
    print(update_clock())

def time_manager():
    time = 3
    
def get_stage():
    global stage
    return stage

def next_stage():
    global stage
    stage+=1

def saveitem(item):
    global stage, subject, result
    with open('Sub '+ str(subject)+ ' Context ' + str(result) + ' clicks.txt', 'w') as f:
        for it in item:
            f.write(str(it) + ' \n')

def experimental_loop():

    root.after(0, update_clock)
    #root.after(0, get_mouse_pos)
    root.after(0,show_cross)
    root.after(0,hidecursor)
    root.after(1000,hide_cross)
    root.after(1000,show_context)
    root.after(2000,hide_context)
    root.after(2000,show_cross)
    root.after(3000,hide_cross)
    root.after(3000,show_images)
    root.after(4000,hide_images)
    root.after(4000,show_empty)
    root.after(5000,hide_empty)
    root.after(5000,show_go)
    root.after(5500,hide_go)
    root.after(5500,show_decision)
    root.after(6500,hide_decision)
    root.after(6500,showcursor)
    root.after(6500,show_complex)
    root.after(10300,get_complex)
    root.after(10500,hide_complex)
    root.after(10500,show_interesting)
    root.after(15300,get_interesting)
    root.after(15500,hide_interesting)
    root.after(15500,show_rewarding)
    root.after(20300,get_rewarding)
    root.after(20500,hide_rewarding)
    root.after(20500,next_stage)

def run_exp():
    global stage
    button.place_forget()
    title.place_forget()
    exp()

def exp():
    for i in range(1):
        time = i*20510
        rest = (1*20510)
        root.after(time, experimental_loop)
        root.after(rest, breaktime)
        
def motion(event):
    x, y = event.x, event.y
    z = (x, y)
    mouse_pos.append(z)

def leftclick(event):
    global mouse_click
    lclick = 0, event.x, event.y, now
    mouse_click.append(lclick)
    #print "Left clicked at ",lclick
    
    
def rightclick(event):
    global mouse_click
    rclick = 1, event.x, event.y, now
    mouse_click.append(rclick)
    #print "Right clicked at ", rclick

def get_mouse_pos():
    for i in range(90):
        flags, hcursor, (x,y) = win32gui.GetCursorInfo()
        z = (x,y)
        mouse_pos.append(z)
        time.sleep(0.01)

        
button = tk.Button(text="Accept & Start the Experiment", command=run_exp, bg="white")
button.place(relx=0.5, rely=0.8, anchor="center")

root.bind('<Motion>', motion)
root.bind("<Button-1>", leftclick)
root.bind("<Button-3>", rightclick)
root.attributes("-fullscreen", False)

root.mainloop()
```
